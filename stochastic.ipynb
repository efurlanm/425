# Stochastic Gradient Descent (SGD)

*Last updated: 2023-10-06*

- To train ANN, it is common to use the Backpropagation and Stochastic Gradient Descent (SGD) algorithms.
- SGD has been used since at least 1960 for training Adaline (Adaptive Linear Neuron), an early single-layer artificial neural network.
- SGD is an iterative method for optimizing an loss function.
- It can be considered as a stochastic approximation of gradient descent optimization, as it uses an estimate calculated from a randomly selected subset of data (rather than using the actual gradient calculated from the entire data set).
- Especially in optimization problems with multiple dimensions, SGD reduces the computational burden in exchange for a lower convergence rate.
- SGD starts from a random point on a function and travels down its slope in steps until it reaches the lowest point of that function
- Machine learning consider the problem of minimizing an objective function that has the form of a sum: $Q(w)={\frac {1}{n}}\sum _{i=1}^{n}Q_{i}(w)$ .
- Each summand function $Q_{i}$ is typically associated with the $i$-th observation in the data set used for training.
- To economize on the computational cost at every iteration, stochastic gradient descent samples a subset of summand functions at every step.

Gradient

- The gradient of a function of several variables is the vector field $\nabla f$ whose value at a point $p$ is the direction in which the function increases most rapidly from $p$.
