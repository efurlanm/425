- Given a set of M pairs of points ($x_i$, $y_i$)
- Present the straight line that best approximates these points
- Find an example where this line has an angular coefficient very close to zero

```python
import matplotlib.pyplot as plt
from scipy import stats as st
import random as rd
rd.seed()
```

```python
m = 10  # elements
x = [int(rd.random()*100) for _ in range(m)]
y = [int(rd.random()*100) for _ in range(m)]
#print(x, y)
```

```python
slope, intercept, r, p, std_err = st.linregress(x, y)
slope
```

```python
plt.scatter(x, y)
plt.plot(x, list(map(lambda x: slope*x+intercept, x)))
plt.show()
```

```python
m = 1000000  # elements
x = [int(rd.random()*100) for _ in range(m)]
y = [int(rd.random()*100) for _ in range(m)]
#print(x, y)
```

```python
slope, intercept, r, p, std_err = st.linregress(x, y)
slope
```

```python
plt.scatter(x[0:9], y[0:9])
plt.plot(x, list(map(lambda x: slope*x+intercept, x)))
plt.show()
```

## Notes

- A horizontal line shows that the variable x does not depend on the variable y
- If the angular coefficient ~= 0, it means that one variable is not dependent on the other
- Another way to analyze is to look at the correlation coefficient
- One of the things this graph shows is whether one variable is dependent on the other
- The mean y is the intersection point, which corresponds to the mean of the y component of the sample
